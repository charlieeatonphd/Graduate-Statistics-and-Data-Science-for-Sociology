## Add tabula, stata, and R magic packages to the Notebook

```python
import tabula
import ipystata

from ipystata.config import config_stata
config_stata('/Applications/Stata/StataMP.app/Contents/MacOS/StataMP')
%load_ext rpy2.ipython
```

## Import PDF with Tabula 

```python
pdf_path = "https://github.com/chezou/tabula-py/raw/master/tests/resources/data.pdf"

dfauto = tabula.read_pdf(pdf_path, stream=True, pages=1)[0]
```

```python
dfauto = dfauto.rename(columns = {'Unnamed: 0':'model'})
```

```python
dfauto
```

## Pass dataframe to Stata and summarize

```python
%%stata -d dfauto
summarize
```

```python
## Pass dataframe to R & create scatterplot
```

```python
%%R -i dfauto
plot(dfauto$hp, dfauto$mpg)
```

```python
## Import Python Seaborn package and graph scatterhist
```

```python
import numpy as np, pandas as pd; np.random.seed(0)
import seaborn as sns; sns.set(style="white", color_codes=True)
g = sns.jointplot(x="hp", y="mpg", data=dfauto)
```
